```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("/kaggle/input/loan-predication/train_u6lujuX_CVtuZ9i (1).csv")
```

```python
df.head(5)
```

```python
# statistical measures
df.describe()
```

```python
# counting number of missing values
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
# label encoding
df.replace({"Loan_Status":{'N':0,'Y':1}},inplace=True)
```

```python
df['Dependents'].value_counts()
```

```python
# replacing the value of 3+ to 4
df = df.replace(to_replace='3+',value=4)
```

```python
df['Dependents'].value_counts()
```

```python
# education and loan status
sns.countplot(x='Education',hue='Loan_Status',data=df)
```

```python
# marital status and loan status
sns.countplot(x='Married',hue='Loan_Status',data=df)
```

```python
df.replace({'Married':{'No':0,'Yes':1},'Gender':{'Male':1,'Female':0},'Self_Employed':{'No':0,'Yes':1},'Property_Area':{'Rural':0,'Semiurban':1,'Urban':2},'Education':{'Graduate':1,'Not Graduate':0}},inplace=True)
```

```python
df.head()
```

```python
X = df.drop(columns=['Loan_ID','Loan_Status'],axis=1)
y = df['Loan_Status']
```

```python
X
```

```python
y
```

```python
# train test split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1,stratify=y,random_state=42)
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train,y_train)
```

```python
# accuracy score on training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,y_train)
```

```python
training_data_accuracy
```

```python
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,y_test)
```
